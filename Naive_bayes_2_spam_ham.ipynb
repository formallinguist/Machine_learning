```python
# Here I  imported pandas library
import pandas as pd
```

```python
# Here I loaded the dataset
df = pd.read_csv("/content/spam.csv")
```

```python
# Here I am exploring the data
df.head()
```

```python
# Here I am grouping it by category and describing it.
df.groupby('Category').describe()
```

```python
# Converting the category to numerical values by creating a new coloumn called spam.
df['spam'] = df['Category'].apply(lambda x: 1 if x== 'spam' else 0)
```

```python
df.head()
```

```python
# Splitting the data into train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.Message,df.spam, test_size=0.25)
```

```python
# We are using count vectorizer to convert the text to numbers.
#we are using features are equal to number of unique words
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
x_train_count = v.fit_transform(x_train.values)
x_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_count,y_train)
```

```python
emails =[
    'you became president.',
    'LCT alumni from the 2018 intake Nina Hosseinikivanani has shared an internship position at the Luxembourg Institute of Science and Technology.'
]

emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
x_test_count = v.transform(x_test)
model.score(x_test_count,y_test)
```

# Creating a pipe line instead of using above cv steps

```python
# Created a pipe line using sklearn
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
clf.predict(emails)
```
