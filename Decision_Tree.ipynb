```python
import pandas as pd
```

```python
df = pd.read_csv("/content/titanic.csv")
```

```python
df.head()
```

```python
inputs = df.drop(['PassengerId', 'Survived', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis=1)
```

```python
inputs.head()
```

```python
target = df['Survived']
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Pclass = LabelEncoder()
le_sex = LabelEncoder()
le_Age = LabelEncoder()
le_Fare = LabelEncoder()
```

```python
inputs['P_class_n'] = le_Pclass.fit_transform(inputs['Pclass'])
inputs['sex_n'] = le_Pclass.fit_transform(inputs['Sex'])
inputs['Age_n'] = le_Pclass.fit_transform(inputs['Age'])
inputs['Fare_n'] = le_Pclass.fit_transform(inputs['Fare'])
```

```python
inputs.head(10)
```

```python
inputs_n = inputs.drop(['Pclass', 'Sex', 'Age', 'Fare'], axis='columns')
```

```python
inputs_n.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs_n,target, test_size=0.2, random_state=1)
```

```python
len(x_train)
```

```python
len(y_train)
```

```python
len(x_test)
```

```python
len(y_test)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
pred = model.predict(x_test)
cm = confusion_matrix(y_test,pred)
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```
