```python
# Here I  imported pandas library
import pandas as pd
```

```python
# Here I loaded the dataset
df = pd.read_csv("/content/titanic.csv")
```

```python
# Here I am exploring the data
df.head()
```

```python
# Dropping the unnecessary columns from df
inputs = df.drop(['PassengerId', 'Survived', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis=1)
```

```python
# Looking at the data frame of the first 6 rows
inputs.head(6)
```

```python
# Using onehot encoding and dummines for the sex column to convert the text to numbers.
dummies = pd.get_dummies(inputs.Sex)
dummies.head(3)
```

```python
#Concatinating the newly created dummies columns to the input df
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head()
```

```python
# Dropping the sex column
inputs.drop('Sex',axis = 'columns', inplace=True)
```

```python
inputs.head()
```

```python
# Here I am checking if theere are any "Nan" Values
inputs.columns[inputs.isna().any()]
```

```python
# converting the nan values to numerical data by taking the mean values
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head(6)
```

```python
# This is the target variable that has target data
target = df['Survived']
```

```python
target
```

```python
len(inputs)
```

```python
# Importing train test split from sklearn.
from sklearn.model_selection import train_test_split
```

```python
# Splitting the data into 80 and 20
x_train, x_test, y_train, y_test = train_test_split(inputs,target, test_size=0.2, random_state=1)
```

```python
len(x_train)
```

```python
len(y_train)
```

```python
len(x_test)
```

```python
len(y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB() #Bell curve
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_test[:10]
```

```python
model.predict(x_test[:10])
```

```python
model.predict_proba(x_test[:10])
```

```python
from sklearn.metrics import confusion_matrix
```

```python
pred = model.predict(x_test)
cm = confusion_matrix(y_test,pred)
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```
